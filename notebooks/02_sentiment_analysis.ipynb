```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
# Load the sentiment analysis results
df = pd.read_csv("C:/project/Twitter-US-Airline-Sentiment/data/processed/reddit_posts_with_sentiment.csv")
```

```python
# 1. Analyze Sentiment Distribution
sentiment_counts = df["sentiment"].value_counts()
sentiment_percentages = sentiment_counts / len(df) * 100

print("Sentiment Distribution:")
print(sentiment_counts)
print("\nSentiment Percentages:")
print(sentiment_percentages)
```

```python
# 2. Visualize Sentiment Distribution
# Create the reports directory if it doesn't exist
import os
os.makedirs("reports", exist_ok=True)

# Plot the sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind="bar", color=["blue", "green", "red"])
plt.title("Sentiment Distribution of Reddit Posts")
plt.xlabel("Sentiment")
plt.ylabel("Number of Posts")
plt.xticks(rotation=0)

# Save the plot
plt.savefig("reports/sentiment_distribution.png")
plt.show()
```
